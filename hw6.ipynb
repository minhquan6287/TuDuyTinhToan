#### W6A16 Nhâp vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

```python
n,m = map(int,input().split())
a = []
for i in range(n):
    h = list(map(int,input().split()))
    a.append(h)
for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:>4}",end = " ")
    print()
```

#### W6A20 Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1


```python
def kiemtra(n, k):
    for i in range(len(n)):
        if n[i] == k:
            return i
    return -1

n = list(map(int, input().split()))
k = int(input())

print(kiemtra(n, k))
```

#### W6A21 Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện 

```python
s = input().split()

d = {}

for i in range(len(s)):
    if s[i] in d:
        d[s[i]] += (i,)
    else:
        d[s[i]] = (i,)

print(d)
```

#### W6A23 VIết hàm xóa đi các số <x có trong mảng

```python
def xoa_nho_hon_x(a, x):
    result = []
    for i in a:
        if i >= x:
            result.append(i)
    return result
a = list(map(int,input().split()))
x = int(input())
print(xoa_nho_hon_x(a,x))
```

#### W6A24 Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí trước sau: 0 2 3 2 4 0 

```python
def khong_to_front(a):
    khong = []
    others = []
    
    for x in a:
        if x == 0:
            khong.append(x)
        else:
            others.append(x)
    
    return khong + others
a = list(map(int,input().split()))
print(khong_to_front(a))
```

#### W6A25 Cho dãy chỉ gồm N giá trị từ 0 - 9 . Kiểm tra xem giá trị nào không có trong 

```python
def tim_so_thieu(a):
    thieu = []
    for i in range(10):   # từ 0 đến 9
        if i not in a:
            thieu.append(i)
    return thieu
a = list(map(int,input().split()))
print(tim_so_thieu(a))
```

#### W6A26 Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 

```python
def tong_cuc_dai(a):
    s = 0
    for i in range(1, len(a) - 1):  # bỏ phần tử đầu và cuối
        if a[i] > a[i-1] and a[i] > a[i+1]:
            s += a[i]
    return s
a = list(map(int,input().split()))
print(tong_cuc_dai(a))
```

#### W6A27 Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước)

```python
def vi_tri_nho_hon_x(a, x):
    for i in range(len(a)):
        if a[i] < x:
            print(i, end=" ")
a = list(map(int,input().split()))
x = int(input())
print(vi_tri_nho_hon_x(a,x))
```

#### W6A28 Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm 


```python
def kiemtra(a):
    tang = True
    giam = True

    for i in range(len(a) - 1):
        if a[i] >= a[i+1]:
            tang = False
        if a[i] <= a[i+1]:
            giam = False

    if tang:
        return "Tang"
    if giam:
        return "Giam"
    return "Khong"
a = list(map(int,input().split()))
print(kiemtra(a))
```
