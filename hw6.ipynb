#### W6A16 Nhâp vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

```python
n,m = map(int,input().split())
a = []
for i in range(n):
    h = list(map(int,input().split()))
    a.append(h)
for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:>4}",end = " ")
    print()
```

#### W6A17 Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và duong cheo phu cua mang

```python
n = int(input())
a = []
for i in range(n):
    h = list(map(int,input().split()))
    a.append(h)
for i in range(n):
    print(a[i][i],end = " ")
print()
for i in range(n):
    print(a[i][n-1-i],end =" ")
```

#### W6A18 Nhâ ̣p vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột k (hàng k)

```python
n,m,k = map(int,input().split())
a = []
for i in range(n):
    h = list(map(int,input().split()))
    a.append(h)
sumcot = 0
for j in range(m):
    sumcot += a[j][k]
print(f"{sum(a[k])}")
print(sumcot)

    
```
