#### Bài 2: Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ta xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2chữ số phần thập phân. 



```python
a = float(input(" Nhập chiều rộng "))
b = float(input(" Nhập chiều dài "))
r = a / 2
dien_tich_hinh_chu_nhat = a * b
dien_tich_hinh_tron = 3.14 * ( r**2)
dien_tich_con_lai = dien_tich_hinh_chu_nhat - dien_tich_hinh_tron
print( f" Diện tích trồng cây xung quanh là : {dien_tich_con_lai:.2f}")
```

#### Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 


```python
a, b, c = map(float, input("Nhập 3 số a, b, c").split())
if a  + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    dien_tich = (p*(p - a)*(p-b)*(p-c))**0.5
    print(f"Là tam giác có diện tích: {dien_tich:.1f}")
else:
    print("Không phải 3 cạnh của tam giác")
```
