#### W6A1 Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]


```python
a = list(map(int, input().split()))

result = []

for x in a:
    if x not in result:
        result.append(x)

print(result)
```

#### W6A2 Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu tới vị trí i), giữ nguyên thứ tự


```python
a = list(map(int, input().split()))

result = []
tong = 0

for i in range(len(a)):
    tong += a[i]
    result.append(tong)

print(result)
```

#### W6A3 Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp xếp.)

#### Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp lại, thêm vào list của key đó


```python
s = input().split()
d = {}
for pair in s:
    key, value = pair.split(":")
    value = int(value)
    if key in d:
        d[key].append(value)
    else:
        d[key] = [value]

print(d)
```

#### Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros': ...} (đếm số >0, <0, =0)


```python
a = list(map(int, input().split()))
d = {'positives': 0, 'negatives': 0, 'zeros': 0}
for x in a:
    if x > 0:
        d['positives'] += 1
    elif x < 0:
        d['negatives'] += 1
    else:
        d['zeros'] += 1
print(d)
```

#### Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của key. 

```python
a = input().split()
b = input().split()
d = {}
e = {}
for i in a:
    k, v = i.split(":")
    d[k] = int(v)
for i in b:
    k, v = i.split(":")
    e[k] = int(v)
r = {}
for k in d:
    r[k] = d[k]
for k in e:
    if k in r:
        r[k] += e[k]
    else:
        r[k] = e[k]
print(dict(sorted(r.items())))
```

#### Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.


```python
a = list(map(int, input().split()))
k = int(input())
r = []
for i in range(len(a)):
    for j in range(i+1, len(a)):
        if a[i] + a[j] == k:
            r.append((a[i], a[j]))
r.sort()  
print(r)
```

#### Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất trong số đó.


```python
a = list(map(int,input().split()))
m = 0
for i in a:
    if a.count(i) > m:
        m = a.count(i)
b = []
for i in a:
    if a.count(i) == m and i not in b:
        b.append(i)
print(min(b))
```

#### [Giao của hai danh sách]
Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên. Hãy tìm các
phần tử chung (giao) của hai danh sách này và in ra một list mới chứa các
phần tử chung đó, không trùng lặp.


```python
a = list(map(int, input().split()))
b = list(map(int, input().split()))
c = []
for i in a:
    if i in b and i not in c:
        c.append(i)
print(c)
```
