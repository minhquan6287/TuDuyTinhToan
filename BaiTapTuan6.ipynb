#### W6A1 Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]


```python
a = list(map(int, input().split()))

result = []

for x in a:
    if x not in result:
        result.append(x)

print(result)
```

#### W6A2 Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu tới vị trí i), giữ nguyên thứ tự


```python
a = list(map(int, input().split()))

result = []
tong = 0

for i in range(len(a)):
    tong += a[i]
    result.append(tong)

print(result)
```

#### W6A3 Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp xếp.)

```python
t = tuple(map(int, input().split()))
k = int(input())

n = len(t)
k = k % n  

a = t[k:] + t[:k]

print(a)
```

#### Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp lại, thêm vào list của key đó


```python
s = input().split()
d = {}
for pair in s:
    key, value = pair.split(":")
    value = int(value)
    if key in d:
        d[key].append(value)
    else:
        d[key] = [value]

print(d)
```

#### Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros': ...} (đếm số >0, <0, =0)


```python
a = list(map(int, input().split()))
d = {'positives': 0, 'negatives': 0, 'zeros': 0}
for x in a:
    if x > 0:
        d['positives'] += 1
    elif x < 0:
        d['negatives'] += 1
    else:
        d['zeros'] += 1
print(d)
```
