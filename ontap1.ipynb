#### 2 Yêu cầu: Cho số nguyên x, trả về True nếu tồn tại số nguyên n ≥ 1 sao cho x = 1 + 2 + ... + n, ngược lại trả về False. Khuyến khích dùng vòng lặp đơn giản để cộng dồn tổng. 


```python
def tontain(x):
    s = 0
    for i in range(1,x+1):
        s += i
        if s == x:
            return True
    return False
print(tontain(6))
```

```python
def taotamgiac(x1,y1,x2,y2,x3,y3):
    ABx = x2 - x1
    ABy = y2 - y1
    ACx = x3 - x1
    ACy = y3 - y1
    if ACx == 0 and ACy ==0:
        return False
    if ACx == 0:
        return ABx != 0
    if ACy == 0:
        return ABy != 0
    if ABx/ACx == ABy/ACy:
        return False
    return True
print(taotamgiac(0,0,1,0,0,1))
```
