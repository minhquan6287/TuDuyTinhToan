#### W5A16 Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số

```python
def is_amstrong(n):
    n = abs(n)
    count = 0
    temp = n 
    total = 0
    if temp == 0:
        count = 1
    else:
        while temp > 0:
            count += 1
            temp //= 10
    temp = n
    for i in range(count):
        a = temp%10
        temp//=10
        total += a**count
    return n == total
print(is_amstrong(153))
    

```

#### W5A17 Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số 


```python
def is_perfect_square(n):
    if n < 0:
        return False
    for i in range(n+1):
        if i*i ==n:
            return True
    return False
print(is_perfect_square(9))
    
```

#### W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.


```python
def so_doi_xung(n):
    if n<0:
        return False
    temp = n
    so_dao_nguoc = 0
    while temp>0:
        a = temp%10
        so_dao_nguoc = so_dao_nguoc*10 +a
        temp//=10
    return n == so_dao_nguoc
print(so_doi_xung(111))
    
```

#### W5A19 VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.

#### W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'

```python
def detect_number_type(s):
    if s.lstrip("-").isdigit():
        return "int"
    if s.count(".") == 1:
        left,right = s.split(".")
        if left.startswith("-"):
            left = left[1:]
        if left.isdigit() and right.isdigit():
            return "float"
    return "neither"
print(detect_number_type("-9.6"))
print(detect_number_type("-9"))
print(detect_number_type("jefgw"))
    
```
