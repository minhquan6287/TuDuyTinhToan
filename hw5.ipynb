#### W5A16 Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số

```python
def is_amstrong(n):
    n = abs(n)
    count = 0
    temp = n 
    total = 0
    if temp == 0:
        count = 1
    else:
        while temp > 0:
            count += 1
            temp //= 10
    temp = n
    for i in range(count):
        a = temp%10
        temp//=10
        total += a**count
    return n == total
print(is_amstrong(153))
    

```

#### W5A17 Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số 


```python
def is_perfect_square(n):
    if n < 0:
        return False
    for i in range(n+1):
        if i*i ==n:
            return True
    return False
print(is_perfect_square(9))
    
```

#### W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.


```python
def so_doi_xung(n):
    if n<0:
        return False
    temp = n
    so_dao_nguoc = 0
    while temp>0:
        a = temp%10
        so_dao_nguoc = so_dao_nguoc*10 +a
        temp//=10
    return n == so_dao_nguoc
print(so_doi_xung(111))
    
```

#### W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'

```python
def detect_number_type(s):
    if s.lstrip("-").isdigit():
        return "int"
    if s.count(".") == 1:
        left,right = s.split(".")
        if left.startswith("-"):
            left = left[1:]
        if left.isdigit() and right.isdigit():
            return "float"
    return "neither"
print(detect_number_type("-9.6"))
print(detect_number_type("-9"))
print(detect_number_type("jefgw"))
    
```

#### W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.


```python
def are_collinear(x1,y1,x2,y2,x3,y3):
    return 1/2 * abs(x1*(y2-y3)) + x2*(y3-y1)+ x3*(y1-y2) == 0
print(are_collinear(1,0,2,0,4,0))
```

#### W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).


```python
def goc(x1,y1,x2,y2,x3,y3):
    import math
    tu = (x1 - x2)*(x3 - x2) + (y1 - y2)*(y3 - y2)
    mau = (((x1 - x2)**2 + (y1 - y2)**2)**0.5) * (((x3 - x2)**2 + (y3 - y2)**2)**0.5)
    return math.acos(tu/mau) *180/(math.pi)

print(goc(1,0,0,0,0,1))
```

#### W5A30 Viết hàm tính hệ số nhị thức C(n, k).

#### W5A33
Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa, 
Các quy tắc cần cài đặt:
1. Giá mở cửa (0–0.8 km):
- Xe 4 chỗ: 11,000 VND
- Xe 7 chỗ: 12,000 VND
2. Giá theo km sau 0.8 km:
- Xe 4 chỗ: từ 0.8→20 km ơnh 13,000 VND/km; trên 20 km tính 10,500 VND/km
- Xe 7 chỗ: từ 0.8→20 km ơnh 14,000 VND/km; trên 20 km tính 12,000 VND/km
3. Phí chờ: 5 phút đầu miễn phí; sau đó, mỗi 5 phút (làm tròn lên theo từng block 5 phút) tính 2,000 VND.
4. Phụ thu ban đêm: nếu start_hour thuộc [22,23,0,1,2,3,4,5], cộng +20% trên tiền tạm tính (mở cửa + quãng đường +
chờ).
5. Điều kiện đặc biệt:
- Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền.
- Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu)

```python
import math

def smart_taxi_fare(km, wait_min, car_type, start_hour):
    if car_type == 4:
        fare = 11000
    else:
        fare = 12000

    if km > 0.8:
        extra_km = km - 0.8
        
        if car_type == 4:
            if km <= 20:
                fare += extra_km * 13000
            else:
                fare += (20 - 0.8)*13000 + (km - 20)*10500
        else:
            if km <= 20:
                fare += extra_km * 14000
            else:
                fare += (20 - 0.8)*14000 + (km - 20)*12000

    if wait_min > 5:
        blocks = math.ceil((wait_min - 5) / 5)
        fare += blocks * 2000

    if start_hour in [22,23,0,1,2,3,4,5]:
        fare *= 1.2

    if km > 30 and start_hour not in [22,23,0,1,2,3,4,5]:
        fare *= 0.9

    if fare < 50000 and km > 5:
        fare = 50000

    return int(fare)
print(smart_taxi_fare(9,7,4,22))
```

#### W5A34
Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau 

```python
def sum_1_to_n(n):
    s = 0
    for i in range(n+1):
        if len(str(i)) == len(set(str(i))):
            s += i
    return s
print(sum_1_to_n(20))
```
