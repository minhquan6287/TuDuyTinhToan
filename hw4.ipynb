#### Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b 

```python
a = int(input())
b = int(input())

for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        print(i, end=" ")
```

#### Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 

```python
n = int(input())

if n <= 0:
    print("False")
else:
    while n % 2 == 0:
        n //= 2
    if n == 1:
        print("True")
    else:
        print("False")
```

#### Nhập vào số nguyên dương n, tính tổng các chữ số của n 


```python
n = int(input())

tong = 0
while n > 0:
    tong += n % 10
    n //= 10

print(tong)
```

#### Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = input()

chan = 0
le = 0

for ch in n:
    if int(ch) % 2 == 0:
        chan += 1
    else:
        le += 1

print("So chu so chan:", chan)
print("So chu so le:", le)
```

#### Nhập n Cho S(k) = 1 + 2 + 3 + … + k Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
n = int(input())

S = 0
k = 0

while S + (k + 1) < n:
    k += 1
    S += k

print(k)
```

#### Nhập vào A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
A = float(input())

tong = 0
n = 0

while tong <= A:
    n += 1
    tong += 1 / n

print(n)
```

#### Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 


```python
a = []

while True:
    x = int(input())
    if x == -1:
        break
    a.append(x)

if len(a) == 0:
    print("Khong co du lieu")
else:
    print("So lon nhat:", max(a))
    print("So nho nhat:", min(a))
```

#### Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
A = int(input())

a, b = 1, 1

while b <= A:
    a, b = b, a + b

print(a)
```

#### Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng) Dùng cách: for trên index; KHÔNG dùng hàm split(), len()

```python
s = input()

count = 0

for i in range(0, len(s)):
    if s[i] != ' ':
        if i == 0 or s[i-1] == ' ':
            count += 1

print(count)
```

#### Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi Dùng cách: for trên index; KHÔNG dùng hàm split()

```python
s = input()

start = -1

for i in range(0, len(s)):
    if s[i] != ' ':
        start = i
        break

if start != -1:
    for i in range(start, len(s)):
        if s[i] == ' ':
            break
        print(s[i], end="")
```

#### Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in thường, bao nhiêu ký tự số 

```python
s = input()

hoa = 0
thuong = 0
so = 0

for i in range(0, len(s)):
    if 'A' <= s[i] <= 'Z':
        hoa += 1
    elif 'a' <= s[i] <= 'z':
        thuong += 1
    elif '0' <= s[i] <= '9':
        so += 1

print("So ky tu in hoa:", hoa)
print("So ky tu in thuong:", thuong)
print("So ky tu so:", so)
```
